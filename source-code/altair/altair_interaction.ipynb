# Altair interaction test

```python
import altair as alt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-6.0*np.pi, 6.0*np.pi, 101)
df = pd.DataFrame({
    'x': x,
    'sin(x)': np.sin(x),
    'cos(x)': np.cos(x),
    'tan(x)': np.tan(x),
})
```

```python
df_melt = df.melt('x').rename(columns={'variable': 'function', 'value': 'y'})
```

```python
df_melt.head()
```

```python
selection = alt.selection_point(nearest=True, on='mouseover',fields=['x', 'y'], empty='none')
legend_selection = alt.selection_point(fields=['function'], bind='legend')
points = alt.Chart(df_melt).mark_point(size=1, filled=True).encode(
    x='x:Q',
    y='y:Q',
    color=alt.Color('function:N'),
    opacity=alt.condition(legend_selection, alt.value(1.0), alt.value(0.2)),    
    tooltip=['x', 'y', 'function']
)
lines = points.mark_line().encode(
    x='x:Q',
    y='y:Q',
    color='function:N',
    opacity=alt.condition(legend_selection, alt.value(1.0), alt.value(0.2)),    
)
alt.layer(points, lines).add_params(selection).add_params(legend_selection).interactive()
```
